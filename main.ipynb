```python
import os
from dotenv import load_dotenv
from openai import OpenAI
from pydantic import BaseModel, Field
from typing import List
import requests
import asyncio
load_dotenv()
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
# os.environ["OPENAI_API_KEY"] = "openaikey"
# client = OpenAI()
```

```python
class Recipe(BaseModel):
    title: str
    ingredients: List[str]
    cooking_time: str
    instructions: List[str]
```

```python
def recipe_generator(dish_name: str):
    
    prompt = f"""
    Generate a detail_recipe for {dish_name}:
    Return a JSON with:
    title,
    ingredients (list),
    cooking_time,
    instructions (list)
    """

    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "user", "content": prompt}
        ],
        response_format={"type": "json_object"}
    )

    recipe_json = response.choices[0].message.content
    recipe = Recipe.model_validate_json(recipe_json)
    return recipe

print(recipe_generator("Chicken Biryani"))
```

```python
class essay_generator(BaseModel):
    topic: str
    word_count: int

    def generate_essay(topic_name: str):

        prompt = f"""
        Generate an essay on {topic_name} 
        Return a JSON with:
        topic,
        word_count,
        essay
        """

        response = client.chat.completions.create(
            model="gpt-4o-mini",
            messages=[
                {"role": "user", "content": prompt}
            ]
        )

        essay = response.choices[0].message.content
        return essay
    
print(essay_generator.generate_essay("The impact of technology on education"))
```

```python
class dress_suggester(BaseModel):
    occasion: str
    weather: str

    def suggest_dress(occasion: str, weather: str):

        prompt = f"""
        Suggest a dress for {occasion} in {weather} weather.
        Return a JSON with:
        occasion,
        weather,
        dress_suggestion
        """

        response = client.chat.completions.create(
            model="gpt-4o-mini",
            messages=[
                {"role": "user", "content": prompt}
            ]
        )

        suggestion = response.choices[0].message.content
        return suggestion
    
print(dress_suggester.suggest_dress("wedding", "hot"))
```

```python
class job_description_generator(BaseModel):
    job_title: str
    industry: str

    def generate_job_description(job_title: str, industry: str):

        prompt = f"""
        Generate a job description for {job_title} in the {industry} industry.
        Return a JSON with:
        job_title,
        industry,
        job_description
        """

        response = client.chat.completions.create(
            model="gpt-4o-mini",
            messages=[
                {"role": "user", "content": prompt}
            ]
        )

        job_description = response.choices[0].message.content
        return job_description
    
print(job_description_generator.generate_job_description("Python Developer", "Technology"))
```

```python
class skill_per_job(BaseModel):
    job_title: str
    industry: str

    def suggest_skills(job_title: str, industry: str):

        prompt = f"""
        Suggest skills for {job_title} in the {industry} industry.
        Return a JSON with:
        job_title,
        industry,
        skills (list)
        """

        response = client.chat.completions.create(
            model="gpt-4o-mini",
            messages=[
                {"role": "user", "content": prompt}
            ]
        )

        skills = response.choices[0].message.content
        return skills
    
print(skill_per_job.suggest_skills("Python Developer", "Technology"))
```

```python


def web_research_agent(query: str):

    url= "https://api.duckduckgo.com/"
    params = {
            "q": query,
            "format": "json"
        }

    response = requests.get(url, params=params)
    data = response.json()

    abstract = data.get("Abstract", "No abstract available")

    return abstract
    
my_research = web_research_agent("What is artificial intelligence?")

print(my_research)
```

```python
def temperature_api(city: str):
    """
    Mock temperature API tool
    """

    mock_data = {
        "New York": "15°C",
        "Los Angeles": "20°C",
        "Chicago": "10°C",
        "Houston": "25°C",
        "Phoenix": "30°C"
    }

    return mock_data.get(city, "City not found")

print(temperature_api("Los Angeles"))
```

```python
def weather_agent(city: str):

    temperature = temperature_api(city)

    return f"The current temperature in {city} is {temperature}"

print(weather_agent("New York"))

  
```
